**Computao Distribuda e Paralelismo:**

*Computao Distribuda*  o processamento dos dados a partir de inmeras mquinas 
para dividir e completar tarefas mais rapidamente. J o *paralelismo* se refere  
execuo de vrias tarefas de forma simultnea  num nico sistema ou entre sistemas.

```python
#Exemplo de multithreading em Python:

import threading
import time

def tarefa(nome):
    print(f"{nome} iniciando")
    time.sleep(2)
    print(f"{nome} finalizando")

#Threads para duas tarefas:
tarefa1=threading.Thread(target=tarefa,args=('Tarefa 1',))
tarefa2=threading.Thread(target=tarefa,args=('Tarefa 2',))

#Inicializao do thread:
tarefa1.start()
tarefa2.start()

#Finalizao do thread:
tarefa1.join()
tarefa2.join()

print("Processamento finalizado.")
```
